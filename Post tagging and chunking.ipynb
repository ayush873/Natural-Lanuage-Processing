```python
import sys
import requests
import bs4
import random
import re
```

```python
def scrapeWikiArticle(url):
    response = requests.get(
        url=url
    )
    text=""
    soup = bs4.BeautifulSoup(response.content, 'html.parser')

    title = soup.find(id="firstHeading").getText()
    content = soup.find_all('p')
    text=text+title+" "
    
    for i in content:
        text=text+i.getText()

    return text

text=scrapeWikiArticle("https://en.wikipedia.org/wiki/COVID-19")
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
text_corr = re.sub(r'\[.*?\]+', '', text)
text_corr = text_corr.replace('\n', '')
```

```python
def Diff(li1, li2):
    li_dif = []
    for i in li1:
        if(i not in li2):
            li_dif.append(i)
    return li_dif
```

```python
words=word_tokenize(text)
print("No of words tokenized: ", len(words))
```

```python
#this step will remove the "(",")",",","." from the list
tokens=[]
for i in words:
    res=re.sub(r'[^\w\s]' , "" ,i)
    if(res!=""):
        tokens.append(res)
    
```

#### Now, we have correct data to apply POS Tagging

```python
nltk.download('averaged_perceptron_tagger')
```

```python
tagged = nltk.pos_tag(tokens)
```

```python
print(tagged)
```

```python
nounTagged=[]
for i in (tagged):
    if(i[1] in ("NN","NNS","NNP","NNPS")):
        nounTagged.append(i)
```

```python
import io
with io.open("noun.txt", "w", encoding="utf-8") as output:
    for i in nounTagged:
        output.write(str(i) + '\n')
```

```python
print("Total noun terms in the data sets are ",len(nounTagged))
```

```python
verbTagged=[]
for i in (tagged):
    if(i[1] in ("VB","VBD","VBN","VBP","VBZ")):
        verbTagged.append(i)
```

```python
with io.open("verb.txt", "w", encoding="utf-8") as output:
    for i in verbTagged:
        output.write(str(i) + '\n')
```

```python
grammer="""NP : {<DT>?<JJ>*<NN>}"""
chunkParser = nltk.RegexpParser(grammer)
tree=chunkParser.parse(tagged)
```

```python
with io.open("NounPhrases.txt", "w", encoding="utf-8") as output:
    for subtree in tree.subtrees(filter=lambda t: t.label() == 'NP'):
        output.write(str(subtree) + '\n')
```
